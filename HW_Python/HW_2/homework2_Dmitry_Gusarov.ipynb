# Задание 1

На маркетплейсе есть огромный объем данных содержащий записи о продажах различных продуктов. Наша цель — рассчитать общий доход, полученный от каждого продукта за определённый период.

Необходимо решить задачу через классический python, numpy и numpy с использованием векторизации


Данные:
products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны'] (при решении через numpy необходимо представить это в виде np.array() )
prices = [20000, 3000, 45600, 5000] (при решении через numpy необходимо представить это в виде np.array() )

Данные о продажах необходимо сгенерировать самостоятельно и вывести в одном параграфов в ноутбуке, ниже приведен пример генерации случайных значений

Данные в различных реализациях должны совпадать.

После генерации необходимо сравнить доход от каждого продукта и время на выполнение.

### 1.1. Классический Python

```python
start = timeit.default_timer()


variant = 4
days = 100000
sales = [random.randint(0, variant+100) for i in range(days)]

products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны']
prices = [20000, 3000, 45600, 5000]

profit_full = [sum([i * j for j in sales]) for i in prices]

print(f"===Прибыль от продаж за {days} дней===")

result = dict(zip(products, profit_full))

for i in result:
    print(f"{i}: прибыль {result[i]}")


elapsed_time_1 = timeit.default_timer() - start
print('\n\nВремя работы программы "Классический Python": {} sec.'.format(elapsed_time_1))
```

### 1.2. Numpy без векторизации

```python
start = timeit.default_timer()


variant = 4
days = 100000
sales = np.array([random.randint(0, variant+100) for i in range(days)])

products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны']
prices = np.array([20000, 3000, 45600, 5000])

profit_full = [sum([i * j for j in sales]) for i in prices]

print(f"===Прибыль от продаж за {days} дней===")

result = dict(zip(products, profit_full))

for i in result:
    print(f"{i}: {result[i]}")


elapsed_time_2 = timeit.default_timer() - start
print('\n\nВремя работы программы "Numpy без векторизации": {} sec.'.format(elapsed_time_2))
```

### 1.3. Numpy с векторизацией

```python
start = timeit.default_timer()


variant = 4
days = 100000
sales = np.array([random.randint(0, variant+100) for i in range(days)])

products = ['автомобиль HAVAL', 'обувь', 'одежда', 'телефоны']
prices = np.array([20000, 3000, 45600, 5000])

prices = prices.reshape(1, len(prices))
sales = sales.reshape(len(sales), 1)

profit_per_day = sales @ prices
profit_full = np.sum(profit_per_day, axis=0)

print(f"===Прибыль от продаж за {days} дней===")

result = dict(zip(products, profit_full))

for i in result:
    print(f"{i}: {result[i]}")


elapsed_time_3 = timeit.default_timer() - start
print('\n\nВремя работы программы "Numpy с векторизацией": {} sec.'.format(elapsed_time_3))
```

### Итоги задания 1:

```python
res_dict = {'Классический Python': elapsed_time_1,
            'Numpy без векторизации': elapsed_time_2,
            'Numpy с векторизацией': elapsed_time_3}

for i in res_dict:
  print(i, res_dict[i], 'sec.')
```

# Задание 2.

Необходимо написать с использованием numpy реализацию одного из следующих алгоритмов:

- heapsort (вариант: 1,5,9,13,17,21)

- mergesort (вариант: 2,6,10,14,18,22)

- stable (вариант: 3,7,11,15,19)

- quicksort (вариант: 4,8,12,16,20)

Сравнить скорость выполнения этих алгоритмов на чистом python, вашей реализации и numpy.sort()

### 2.1. Классический Python

```python
start_1 = timeit.default_timer()


def quicksort(nums):
   if len(nums) <= 1:
       return nums
   else:
       q = random.choice(nums)
   l_nums = [n for n in nums if n < q]

   e_nums = [q] * nums.count(q)
   b_nums = [n for n in nums if n > q]
   return quicksort(l_nums) + e_nums + quicksort(b_nums)


variant = 4
count = 100000
lst_numbers = [random.randint(0, variant+100) for i in range(count)]

res = quicksort(lst_numbers)
print(res)


time_1 = timeit.default_timer() - start_1
print('\n\nВремя работы программы "Классический Python": {} sec.'.format(time_1))
```

### 2.2. Numpy без векторизации

```python
start_2 = timeit.default_timer()


def quicksort(nums):
   if len(nums) <= 1:
       return nums
   else:
       q = random.choice(nums)
   l_nums = [n for n in nums if n < q]

   e_nums = [q] * list(nums).count(q)
   b_nums = [n for n in nums if n > q]
   return quicksort(l_nums) + e_nums + quicksort(b_nums)


variant = 4
count = 100000
lst_numbers = np.array([random.randint(0, variant+100) for i in range(count)])

res = quicksort(lst_numbers)
print(res)


time_2 = timeit.default_timer() - start_2
print('\n\nВремя работы программы "Numpy без векторизации": {} sec.'.format(time_2))
```

### 2.3. Numpy с векторизацией

```python
start_3 = timeit.default_timer()


variant = 4
count = 100000
lst_numbers = np.array([random.randint(0, variant+100) for i in range(count)])

res = np.sort(lst_numbers)
print(res)


time_3 = timeit.default_timer() - start_3
print('\n\nВремя работы программы "Numpy с векторизацией": {} sec.'.format(time_3))
```

```python
data = np.genfromtxt("https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv",
                     delimiter=',', dtype=None, names=True, encoding=None)

for i, column_name in enumerate(data.dtype.names[1:]):
    record = data[np.argmax(data[column_name])]
    print(f"Наибольшее потребление {column_name}: в стране {record[0]}")
    print(f"Количество: {record[i+1]}\n")
```

<h1>Задание 4</h1>

У нас есть Полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Давайте представим эти данные в виде структурированных массивов и определим:

- число проклятий, летальных исходов и их отношение по фильмам

- частота употребления конкретных прокрятий

- распределение времени между проклятиями

```python
data = np.genfromtxt("https://raw.githubusercontent.com/fivethirtyeight/data/master/tarantino/tarantino.csv",
                     delimiter=',', dtype=None, names=True, encoding=None)

movies = np.unique(data['movie'])
all_statistic = np.zeros((len(movies), 4))

for i, movie in enumerate(movies):
    current_movie = data[data['movie'] == movie]
    curse = len(current_movie[current_movie['type'] == 'word'])
    death = len(current_movie[current_movie['type'] == 'death'])
    all_statistic[i] = [curse, death, 0, 0]

all_curse = np.sum(all_statistic[:, 0])
all_death = np.sum(all_statistic[:, 1])
all_statistic[:, 2] = all_statistic[:, 0] / all_curse
all_statistic[:, 3] = all_statistic[:, 1] / all_death

np.set_printoptions(suppress=True)

for pos in range(len(movies)):
    stat = all_statistic[pos]
    print(f"Фильм: {movies[pos]}".format())
    print(f"Количество проклятий: {int(stat[0])} и их отношение к общему количеству: {stat[2]}")
    print(f"Количество смертей: {int(stat[1])} и их отношение к общему количеству: {stat[3]}\n")
```
